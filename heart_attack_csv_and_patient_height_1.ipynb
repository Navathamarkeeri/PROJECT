```python
# remove outliers based on age column

Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]
```

```python
import pandas as pd

# Load your data into the 'df' DataFrame
df = pd.read_csv('/content/Heart_Attack_Risk_Updated.csv') # Replace 'your_data.csv' with your actual file path

def apply_function(row):
    if row['Age'] > 50 and row['Cholesterol'] <= 200:
        return row['Age'] * 2
    elif row['Age'] > 50 and row['Cholesterol'] > 200:
        return row['Age'] / 2
    else:
        return 0

# Apply the function and create the 'New_Column'
df['New_Column'] = df.apply(apply_function, axis=1)

# Display the updated DataFrame
df.head()
```
