```python
import pandas as pd
```

```python
df = pd.read_excel('/content/Length_of_Stay_Dataset.xlsx')
```

```python
df.head()
```

```python
df.head(50)
```

```python
total_patients = df['Patient ID'] .nunique()
print("total patients:", total_patients)
```

```python
average_los = df['Length of Stay in days'].mean()
print(f"Average length of stay: {average_los} days")
```

```python
admission_type = df['Admission Type'].unique()
print(f"Admission Type: {admission_type} ")
```

```python
primary_diagnosis = df['Primary Diagnosis'].value_counts()
print(f"Primary Diagnosis:{primary_diagnosis} ")
```

```python
avg_age = df.groupby('Admission Type')['Age'].mean()
print(avg_age)
```

```python
critical_patients = df[(df['Severity of Illness'] == 'Critical') & (df['Length of Stay in days'] > 30)]
print("number of critical patients", len(critical_patients) )
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Length of Stay in days'], bins=15, edgecolor='black')
plt.title('Distribution of Length of Stay')
plt.xlabel('Length of Stay (days)')
plt.ylabel('Frequency')
plt.show()
```

```python
admission_type_counts = df['Admission Type'].value_counts()

admission_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(8, 8), legend=True)
plt.title('Distribution of Patients by Admission Type')
plt.ylabel('')  # Remove the y-label for cleaner visualization
plt.show()
```

```python
admission_type_counts = df['Admission Type'].value_counts()

admission_type_counts.plot(kind='bar', figsize=(8, 5), edgecolor='black')
plt.title('Number of Patients by Admission Type')
plt.xlabel('Admission Type')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Age'], df['Length of Stay in days'], alpha=0.7)
plt.title('Age vs Length of Stay')
plt.xlabel('Age')
plt.ylabel('Length of Stay (days)')
plt.show()
```
